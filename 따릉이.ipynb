```python
result = df['대여일자'].value_counts().sort_index().to_frame()
answer = result[result.대여일자  == result.대여일자.max()].index[0]

display(result)
print(answer)
```

연령대 코드가 20대인 데이터를 추출하고,이동거리값이 추출한 데이터의 이동거리값의 평균 이상인 데이터를 추출한다.최종 추출된 데이터를 대여일자, 대여소 번호 순서로 내림차순 정렬 후 1행부터 200행까지의 탄소량의 평균을 소숫점 3째 자리까지 구하여라

```python
#condition1 : 20대
#condition2 : 이동거리값 추출
```

```python
df.head()
```

```python
tw = df[df.연령대코드 =='20대'].reset_index(drop=True)
tw_mean = tw[tw.이동거리 >= tw.이동거리.mean()].reset_index(drop=True)
tw_mean['탄소량'] =tw_mean['탄소량'].astype('float')
target =tw_mean.sort_values(['대여일자','대여소번호'], ascending=False).reset_index(drop=True).iloc[:200].탄소량
result = round(target.sum()/len(target),3)
print(result)
```

```python
condition1 = df[df['연령대코드']=='20대'].reset_index(drop=True)
condition1
```

```python
#이동거리값이 추출한 데이터의 이동거리값의 평균 이상인 데이터를 추출
condition2 = condition1[condition1.이동거리 >= condition1.이동거리.mean()].reset_index(drop=True)
condition2
```

```python
 #최종 추출된 데이터를 대여일자,대여소 번호 순서로 내림차순 정렬 후
condition2['탄소량'] = condition2['탄소량'].astype('float')
```

```python
condition2
```

```python
# target =tw_mean.sort_values(['대여일자','대여소번호'], ascending=False).reset_index(drop=True).iloc[:200].탄소량
condition2 = condition2.sort_values(['대여일자', '대여소번호'],ascending = False)
```

```python
# 1행부터 200행까지의 탄소량의 평균을 소숫점 3째 자리까지 구gkrl ###;;;
len(condition2.iloc[1:200]['탄소량'])
```
